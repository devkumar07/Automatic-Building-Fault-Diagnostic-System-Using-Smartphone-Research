```python
import pandas as pd
import numpy as np

data = pd.read_csv('data/result.csv',parse_dates=['time'], index_col=['time'])
sample = data.loc['2017-07-04 07:30:00':'2017-07-04 08:30:00']
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

plt.rcParams.update({'font.size': 32})
fig, ax = plt.subplots()
hours = mdates.HourLocator(interval = 1)
h_fmt = mdates.DateFormatter('%H:%M:%S')

ax.plot(sample.index, sample['zone_temp'].values, color = 'red', linewidth = 5, label = 'zone temp')
ax.plot(sample.index, sample['supply_temp'].values, color = 'green', linewidth = 5, label = 'supply temp')
ax.plot(sample.index, sample['occupancy'].values, color = 'black', linewidth = 5, label = 'occupancy')
ax.plot(sample.index, sample['outdoor_temp'].values, color = 'violet', linewidth = 5, label = 'outdoor temp')
#Then tick and format with matplotlib:
ax.xaxis.set_major_locator(hours)
ax.xaxis.set_major_formatter(h_fmt)

fig.autofmt_xdate()
fig.set_size_inches(18, 12)
plt.legend()
plt.show()
```
