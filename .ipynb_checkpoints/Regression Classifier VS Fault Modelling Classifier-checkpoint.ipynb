# Simulating fault for regression model

```python
##Faults:
# 1. Offset Fault: x' = a + x + err
# 2. Gain Fault: x' = Bx+ err
# 3. Stuck at Fault: x' = a
# 4. Out-of-bound: x'<theta1 and x'>theta2

def simulateFault(sensedValues, att, params):
    end_fault_index = int(len(sensedValues) * 0.5)
    interval = end_fault_index/4
    BMS = []
    true_values = []
    BMS_final = []
    classification = []
    
    noise = random.rand(len(sensedValues[att])) * 0.5

    for i in range(0, len(sensedValues)):
        if i <= interval:
            a = params['constant'] 
            sensedValues[att][i] = noise[i] + a + sensedValues[att][i]
            classification.append(1)

        elif i > interval and i <= interval*2:
            B = params['coeff']
            sensedValues[att][i] = noise[i] + B * sensedValues[att][i]
            classification.append(1)

        elif i > interval*2 and i<= interval*3:
            a = params['stuck']
            sensedValues[att][i] = a 
            classification.append(1)
            
        elif i > interval*3 and i<= interval*4:
            type_fault = random.randint(1,3)
            low_bound = params['bounds'][0]
            up_bound = params['bounds'][1]
            if type_fault == 1:
                sensedValues[att][i] = low_bound - noise[i] 
            else:
                #print(up_bound + noise[i])
                sensedValues[att][i] = up_bound + noise[i]
            classification.append(1)
        else:
            sensedValues[att][i] = sensedValues[att][i] + noise[i]
            classification.append(0)
    return sensedValues, classification
```

# Classification using SVM by mathematical modelling of Faults

```python
data = pd.read_csv('data/result.csv', parse_dates=['time'], index_col=['time'])


train_data = data.loc['2017-07-10 00:00:00':'2017-07-14 22:55:00']
test_data1 = data.loc['2017-07-17 08:00:00':'2017-07-17 18:00:00']
test_data2 = data.loc['2017-07-18 08:00:00':'2017-07-18 18:00:00']
test_data3 = data.loc['2017-07-19 08:00:00':'2017-07-19 18:00:00']
test_data4 = data.loc['2017-07-20 08:00:00':'2017-07-20 18:00:00']
test_data5 = data.loc['2017-07-21 08:00:00':'2017-07-21 18:00:00']


# train_data = data.loc['2017-07-03 00:00:00':'2017-07-07 22:55:00']
# test_data1 = data.loc['2017-07-10 08:00:00':'2017-07-10 18:00:00']
# test_data2 = data.loc['2017-07-11 08:00:00':'2017-07-11 18:00:00']
# test_data3 = data.loc['2017-07-12 08:00:00':'2017-07-12 18:00:00']
# test_data4 = data.loc['2017-07-13 08:00:00':'2017-07-13 18:00:00']
# test_data5 = data.loc['2017-07-14 08:00:00':'2017-07-14 18:00:00']
#data = data[52992:56992]
#print(test_data3)
```

```python
##Faults:
# 1. Offset Fault: x' = a + x + err
# 2. Gain Fault: x' = Bx+ err
# 3. Stuck at Fault: x' = a
# 4. Out-of-bound: x'<theta1 and x'>theta2

def simulateFault(sensedValues, params):
    end_fault_index = int(len(sensedValues) * 0.5)
    interval = end_fault_index/4
    BMS = []
    true_values = []
    BMS_final = []
    
    noise = random.rand(len(sensedValues)) * 0.5

    for i in range(0, len(sensedValues)):
        if i <= interval:
            a = params['constant']
            BMS.append(noise[i] + a + sensedValues[i])

        elif i > interval and i <= interval*2:
            B = params['coeff']
            BMS.append(noise[i] + B * sensedValues[i])

        elif i > interval*2 and i<= interval*3:
            a = params['stuck']
            BMS.append(a)
            
        elif i > interval*3 and i<= interval*4:
            type_fault = random.randint(1,3)
            low_bound = params['bounds'][0]
            up_bound = params['bounds'][1]
            if type_fault == 1:
                BMS.append(low_bound-noise[i])
            else:
                BMS.append(up_bound + noise[i])
        else:
            BMS.append(noise[i] + sensedValues[i])
    BMS = np.array(BMS)
    sensedValues = np.array(sensedValues)      
    BMS = BMS.round(decimals = 2)
    sensedValues = sensedValues.round(decimals=2)
    for i in range(0, len(sensedValues)-2):
        temp = BMS[i:i+3]
        if i  <= interval:
            BMS_final.append(temp)
            true_values.append(sensedValues[i:i+3])

        elif i > interval and i <= interval*2:
            BMS_final.append(temp)
            true_values.append(sensedValues[i:i+3])


        elif i > interval*2 and i <= interval*3:
            BMS_final.append(temp)
            true_values.append(sensedValues[i:i+3])
            
        elif i > interval*3 and i<= interval*4:
            BMS_final.append(temp)
            true_values.append(sensedValues[i:i+3])
        else:
            BMS_final.append(temp)
            true_values.append(sensedValues[i:i+3])
    return np.array(BMS_final), np.array(true_values)
```

```python
def prepare_data(BMS, sensor_data):
    vectors = []
    end_fault_index = int(len(sensor_data) * 0.5)
    interval = end_fault_index/4
    for i in range(0, len(sensor_data)):
        temp = []
        for j in range(0,len(BMS[i])):
            temp.append([sensor_data[i][j],BMS[i][j]])
        vectors.append(temp)
    return vectors
```

```python
def sample_vectors(vectors, fault_rate):
    result_x = []
    end_fault_index = int(len(vectors) * fault_rate)
    for i in range(0, end_fault_index):
        fault = random.randint(0, int(len(vectors)*0.5))
        result_x.append(vectors[fault])
    for i in range(len(result_x),len(vectors)):
        fault = random.randint(int(len(vectors)*0.5),len(vectors))
        result_x.append(vectors[fault])
        
    return result_x
```

```python
def process_data(vectors, bounds):
    processed_data = []
    classification = []
    for i in range(0,len(vectors)):
#         print(vectors[i])
#         print(bounds[1])
#         print(bounds[0])
        if vectors[i][0][1] == vectors[i][1][1] or vectors[i][1][1] == vectors[i][2][1]:
            processed_data.append([0,1,1])
            classification.append(1)
        elif ((vectors[i][0][1] >= bounds[1] or bounds[0] >= vectors[i][0][1]) and (vectors[i][1][1] >= bounds[1] or bounds[0] >= vectors[i][1][1])) or ((vectors[i][1][1] >= bounds[1] or bounds[0] >= vectors[i][1][1]) and (vectors[i][2][1] >= bounds[1] or bounds[0] >= vectors[i][2][1])):
            processed_data.append([1,0,0])
            classification.append(1)
        elif vectors[i][0][1] - vectors[i][0][0] > 0.5 and vectors[i][1][1] - vectors[i][1][0] > 0.5:
            if vectors[i][0][0] == vectors[i][1][0] or vectors[i][1][0] == vectors[i][2][0] or vectors[i][0][0] == vectors[i][2][0]:
                fault = random.randint(1,3)
                if fault == 1:
#                     print(vectors[i])
#                     print(1)
#                     print('goes here 1')
                    processed_data.append([0,0,1])
                    classification.append(1)
                else:
#                     print(vectors[i])
#                     print(2)
#                     print('goes here 2')
                    processed_data.append([0,1,0])
                    classification.append(1)
            else:
                slope1 = abs((vectors[i][0][1] - vectors[i][1][1])/(vectors[i][0][0]-vectors[i][1][0]))
                slope2 = abs((vectors[i][2][1] - vectors[i][1][1])/(vectors[i][2][0]-vectors[i][1][0]))
                if slope1 > 2 or slope2 > 2:
#                     print(vectors[i])
#                     print(2)
#                     print('goes here 3')
                    processed_data.append([0,1,0])
                    classification.append(1)
                else:
#                     print(vectors[i])
#                     print(1)
#                     print('goes here 4')
                    processed_data.append([0,0,1])
                    classification.append(1)
        elif vectors[i][0][1] - vectors[i][0][0] <= 0.5 and vectors[i][1][1] - vectors[i][1][0] <= 0.5:
            #print(vectors[i])
            classification.append(0)
            processed_data.append([0,0,0])
    return processed_data, classification
```
