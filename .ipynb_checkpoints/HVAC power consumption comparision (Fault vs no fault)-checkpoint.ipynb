```python
import pandas as pd
import numpy as np

data = pd.read_csv('clean.csv')
data['Date'] = pd.to_datetime(data['Date'])
average_no_fault = data.groupby([data['Date'].dt.year,data['Date'].dt.month])['HVAC Power'].mean()
average_no_fault
```

```python
fault_data = pd.read_csv('fault_reduced.csv')
fault_data['Date'] = pd.to_datetime(fault_data['Date'])
average_fault_reduced = fault_data.groupby([fault_data['Date'].dt.year,fault_data['Date'].dt.month])['HVAC Power'].mean()
average_fault_reduced
```

```python
fault_data = pd.read_csv('fault_increased.csv')
fault_data['Date'] = pd.to_datetime(fault_data['Date'])
average_fault_increased = fault_data.groupby([fault_data['Date'].dt.year,fault_data['Date'].dt.month])['HVAC Power'].mean()
average_fault_increased
```

```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 20})
plt.figure(figsize=(20,10))
plt.title('Average HVAC power consumption -- No fault')
plt.bar(['Jan', 'Feb', 'Mar', 'Apr'], average_no_fault)
```

```python
plt.figure(figsize=(20,10))
plt.title('Average HVAC power consumption -- Fault offset reduced by 2 degrees')
plt.bar(['Jan', 'Feb', 'Mar', 'Apr'], average_fault_reduced)
```

```python
plt.figure(figsize=(20,10))
plt.title('Average HVAC power consumption -- Fault offset increased by 2 degrees')
plt.bar(['Jan', 'Feb', 'Mar', 'Apr'], average_fault_increased)
```
